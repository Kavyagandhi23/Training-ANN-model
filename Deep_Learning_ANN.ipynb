Creating Artificial Neural Network (ANN) model using deep learning concepts and tensorflow.

Data use to train the model is https://www.kaggle.com/datasets/shrutimechlearn/churn-modelling

I learned so many new concepts and how to create ANN model and implement practically. Here in this small project, i used churn modelling dataset downloaded from Kaggle. Firstly, I performed Some EDA like drop some columns, convert columns into dummy variable and join the dataframes using concat() function.

Second, Feature Scaling is done by using StandardScaler method, to scale the training and test data in a particular range. Next, the ANN model is created using Sequential model and tensorflow framework. early stopping is also performed to stop the model when the higher accuracy is achieved by the model. Later the Adam optimizer and binary cross entropy is used to calculate the loss and perform the back propogation with the help of Adam optimizer.

Lastly, the prediction is made using the predict() function and the accuracy is calculated from sklearn.models.

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Churn_Modelling.csv")
```

```python
data.head()
```

```python
x_data = data.iloc[:,3:13]
y_data = data.iloc[:,13]
```

```python
x_data.head()
```

```python
y_data
```

```python
Geography = pd.get_dummies(x_data['Geography'])
```

```python
Gender = pd.get_dummies(x_data['Gender'])
```

```python
x_data = x_data.drop(['Geography','Gender'], axis= 1)
```

```python
x_data = pd.concat([x_data, Geography, Gender], axis = 1)
```

```python
x_data
```

```python
# Split the dataset into train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x_data, y_data, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) # fit_transform() used to fit the training data and learn the parameters of the scaling of training data then will use to scale the train data
x_test = sc.transform(x_test)  # transform() use to scale the data learned from the parameters find in the training data
```

```python
# Create ANN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU
from tensorflow.keras.layers import Dropout
classifier = Sequential()
```

```python
x_train.shape
```

```python
# Input Layer
classifier.add(Dense(units=13, activation = 'relu'))
```

```python
# First Hidden Layer
classifier.add(Dense(units = 9, activation = 'relu'))
```

```python
# Second hidden layer
classifier.add(Dense(units = 6, activation = 'relu'))
```

```python
# Output Layer
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate = 0.01)
classifier.compile(optimizer = opt, loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier
```

```python
early_stopping = tensorflow.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model = classifier.fit(x_train,y_train, validation_split = 0.33, batch_size = 10, epochs = 1000, callbacks = early_stopping)
```

```python
model.history.keys()
```

```python
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_pred, y_test)
confusion_matrix
```

```python
from sklearn.metrics import accuracy_score
accuracy_score = accuracy_score(y_pred, y_test)
accuracy_score
```
